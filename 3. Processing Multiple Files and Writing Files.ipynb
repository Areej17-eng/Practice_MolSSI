```python
# Just like previously, we used to define a file path and open the file using with 
# open command, we are gonna use another command here instead of file path.

import glob
glob_file_path = glob.glob(os_file_path)            # file_path contains a list of all files 

for files in glob_file_path:
    with open (files , "r") as outfile:
        data = outfile.readlines()
        for line in data:
            if "Final Energy" in line:
                energy_line = line
                # print(energy_line)
                energy_words = energy_line.split()
                print(energy_words[3])
```

```python
# To get the file names along with their energy data, we can use the following code:

# first_file = file_path[0]
# print(first_file)

first_file = glob_file_path[0]
print(first_file)

first_filename = os.path.basename(first_file)
print(first_filename)

slpit_1_filename = first_filename.split('.')    # This splits into ['propanol', 'out']
molecule_1_filename = slpit_1_filename[0]
print(molecule_1_filename)
```

```python
# Now using the concept above, we can modify our loop so that it prints the file name
# along with each energy value.

import glob
glob_file_path = glob.glob(os_file_path)            # file_path contains a list of all files 

for files in glob_file_path:
    with open (files, "r") as outfiles:            # To print energy values
        data = outfiles.readlines()
        for line in data:
            if "Final Energy" in line:
                energy_line = line
                energy_words = energy_line.split()
                energy_values = energy_words[3]
                #print(energy_values)
    
    filenames = os.path.basename(files)
    split_filenames = filenames.split('.')
    molecule_names = split_filenames[0]
    
    print(molecule_names, energy_values)
```

```python
# Now let's print our our extracted information above in a .txt file.

with open ('energies' , "w+",) as txt_file:

    for files in glob_file_path:
        with open (files, 'r') as outfiles:
            data = outfiles.readlines()
            for line in data:
                if "Final Energy" in line:
                    energy_line = line
                    #print(energy_line)
                    energy_words = energy_line.split()
                    energy_values = energy_words[3]
                    #print(energy_values)

        filenames = os.path.basename(files)
        split_filenames = filenames.split('.')
        molecule_names = split_filenames[0]
        
        print(molecule_names, energy_values)
        txt_file.write(F'For the file {molecule_names} \t the energy is {energy_values} \n')
txt_file.close()
```
