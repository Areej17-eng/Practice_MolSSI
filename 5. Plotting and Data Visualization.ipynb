```python
# Let's use the same file to plot its data:

import os
import numpy

distance_file = os.path.join('CMS Workshp (MolSSI data)' , 'distance_data_headers.csv')
distance_data = numpy.genfromtxt(fname = distance_file, delimiter = ',' , dtype = 'unicode')
header = distance_data[0,:]        # You can also write distance_data[0] to automatically consider all col
data = distance_data[1:,:]         # You can also write distance_data[1:] //     //     //
data = data.astype(numpy.float64)

# Now to plot the data:

import matplotlib.pyplot          # matplotlib is the library and pyplot is the function for plotting
matplotlib.pyplot.figure()        # This initializes a new figure
matplotlib.pyplot.plot(data[:,1])                  # To plot all rows and only second columnn of data



# import matplotlib.pyplot as distance_plot      # matplotlib is the library and pyplot is the function for plotting
# distance_plot.figure()
# distance_plot.plot(data[:,1])                  # To plot all rows and only second columnn of data

```

```python
# To avoid typing matplotlib.pyplot, we can call it plt or anything

import matplotlib.pyplot as plt         
plt.figure()        
plt.plot(data[:,1])            
```

```python
# Now to label the figure and x/y axis:

THR4_ATP = header[1]      # which is basically "THR4_ATP"

plt.figure()
plt.xlabel('Solution Frame')          # To show xlable on the plot 
plt.ylabel('Distance (angstrom)')     # To show ylable on the plot
Figure_1 = plt.plot(data[:,1] , label = THR4_ATP)   
plt.legend()                        # To show legend (plot name) on the plot

# To print out the file:
plt.savefig(F'{THR4_ATP}.png')  # It is just to save the file (not to create a plot)
plt.savefig("THR4_ATP")     # This is the same thing
# Also, here THR4_ATP = header[1] = "THR4_ATP", so the file is saved with "THR4_ATP" name.

# to increase the resolution of our plot (dpi means dots per inches):
plt.savefig(F'{THR4_ATP}.png' , dpi = 300)
```

```python
# Let's plot two plots at the same time:

plt.figure()
plt.xlabel("Solution Frame")
plt.xlabel('Distance (angstrom)')
plt.plot(data[:,1], label = "THR4_ATP")
plt.plot(data[:,2], label = "THR4_ASP")
plt.legend()
plt.savefig("THR4_ATP and THR4_ASP" , dpi = 300)
```

```python
# If I want to plot all the columns, then we can use a for loop. Here, we put the x and y labels and 
# savefig command outside of the for loop since those things only need to be done once.

print(len(header))
for i in range(1, len(header)):      # To generalize, we used range(1,len(header)), its same as range(1,5)
    plt.plot(data[:,i] , label = header[i])
    plt.legend()

plt.xlabel("Solution Frame")
plt.ylabel('Distance (angstrom)')
plt.savefig("all_columns" , dpi = 300)
```

```python
# Now if I want to make separate plots for each column

print(len(header))
for i in range(1, len(header)):      # To generalize, we used range(1,len(header)), its same as range(1,5)
    plt.figure()
    plt.plot(data[:,i] , label = header[i])
    plt.legend()
    plt.xlabel("Solution Frame")
    plt.ylabel('Distance (angstrom)')
    plt.savefig(F'{header[i]}.png' , dpi = 300)
    
```

```python
# If I want to define my own upper and lower limits for each graph:

# If defined out of for loop, then only 1 data_min and 1 data_max value will be generated for all columns.
# data_min = numpy.min(data[:,1:]) * 0.9     # To define my lower limit
# data_max = numpy.max(data[:,1:]) * 1.1     # To define my upper limit

print(len(header))
for i in range(1, len(header)):      # To generalize, we used range(1,len(header)), its same as range(1,5)
    data_min = numpy.min(data[:,i]) * 0.9     # To define my lower limit
    data_max = numpy.max(data[:,i]) * 1.1     # To define my upper limit
    plt.figure()
    plt.plot(data[:,0] , data[:,i] , label = header[i])
    plt.legend()
    plt.ylim(data_min , data_max)
    plt.xlabel("Solution Frame")
    plt.ylabel('Distance (angstrom)')
    plt.savefig(F'New {header[i]}' , dpi = 300)

# Note that in the plt.plot function, we used data[:,0] (the first column) as x-axis and rest as y-axis.
```
