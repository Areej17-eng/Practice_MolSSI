```python
# We are going to use "ethanol.out" file, so to define its relative file path:
import os

ethanol_file = os.path.join('CMS Workshp (MolSSI data)', 'outfiles', 'ethanol.out')
print(ethanol_file)

# As this file is running in "Learning Data" directory, so the file path is defined after that. 
# "ethanol_file" is defined as a relative path to access our file ethanol.out
```

```python
# We are going to use the pre-defined "ethanol_file" to define its absolute path:
absolute_path = os.path.abspath(ethanol_file)
print(absolute_path)
```

```python
# To open the file store at pre-defined "ethanol_file" location

with open(ethanol_file, "r") as opened_ethanol_file:
    file_list = opened_ethanol_file.readlines()
# In this code, we defined the opened file as "opened_ethanol_file", and stored all the lines of
# opened file as a list called "file_list". "r" shows reading only

print(len(file_list))
# Now we are printing the number of lines in our file_list.
```

```python
# Now let's prints all the lines of our "file_list".

for lines in file_list:
    print(lines)
```

```python
# Now we are going to search this entire file_list for "Final Energy" and will print all those values.

for lines in file_list:
    if "Final Energy" in lines:
        energy_lines = lines
print(energy_lines)

## NOW Let's split the words from energy_lines and print energy value (the last word in energy line)
# split
words = energy_lines.split()
print(words)

print('')        # Just to add some enter space among results
energy_value = words[-1]
print(energy_value)

# I can also split like this:
print('')
words = energy_lines.split(':')    # Using : as delimeter (basis of splitting)
print(words)

# we can print energy_value like this:
print('')       
energy_value = words[-1]
print("The Final Energy value is",energy_value)
print(f"The Final Energy value: {energy_value}")
```

```python
## Let's do an exercise for the file "sapt.out" file.

# Opening the file:
import os
sapt_file = os.path.join('CMS Workshp (MolSSI data)', 'sapt.out')
print(sapt_file)

# Or you can say outfile=open(sapt_file,"r")
with open (sapt_file, "r") as outfile: 
    data = outfile.readlines()
    for line in data:
        if "Electrostatics   " in line:
            Estatic_energy = line
            print(Estatic_energy)
        if "Exchange   " in line:
            Exch_energy = line
            print(Exch_energy)
        if "Induction   " in line:
            Induc_energy = line
            print(Induc_energy)
        if "Dispersion   " in line:
            Dispers_energy = line
            print(Dispers_energy)

# Now split the words:
Estatic_words = Estatic_energy.split()
print("Electrostatics:", Estatic_words[3], Estatic_words[4])

Exch_words = Exch_energy.split()
print("Exchange:", Exch_words[3], Exch_words[4])

Induc_words = Induc_energy.split()
print("Induction:", Induc_words[3], Induc_words[4])

Dispers_words = Dispers_energy.split()
print("Dispersion:", Dispers_words[3], Dispers_words[4])

# Now to change them to float type:
Estatic_words[3] = float(Estatic_words[3])
Exch_words[3] = float(Exch_words[3])
Induc_words[3] = float(Induc_words[3])
Dispers_words[3] = float(Dispers_words[3])

Total_energy = Estatic_words[3] + Exch_words[3] + Induc_words[3] + Dispers_words[3]
print("Total_energy:", Total_energy, "[kcal/mol]")
```

```python
## Let's do the same exercise for the file "sapt.out" file by another code.

all_energy_lines = []     # We made a list of all energy lines
all_energy_values = []    # We made a list of all energy values
all_energy_types = []    # We made a list of all energy values

# Opening the file:
import os
sapt_file = os.path.join('CMS Workshp (MolSSI data)', 'sapt.out')
print(sapt_file)

with open (sapt_file, "r") as outfile:
    data = outfile.readlines()
    for line in data:
        if "Electrostatics   " in line:
            Estatic_energy = line
            all_energy_lines.append(Estatic_energy)
        if "Exchange   " in line:
            Exch_energy = line
            all_energy_lines.append(Exch_energy)
        if "Induction   " in line:
            Induc_energy = line
            all_energy_lines.append(Induc_energy)
        if "Dispersion   " in line:
            Dispers_energy = line
            all_energy_lines.append(Dispers_energy)

# Now split the words in "all_energy_lines" list. For a list, we first need to access lines in the list
for line in all_energy_lines:
    words = line.split()
    # print(words)
    energy_type = words[0]
    energy_value = float(words[3])
    all_energy_values.append(energy_value)
    print(f'{energy_type} : {energy_value} kcal/mol')

Total_energy = sum(all_energy_values) # We are adding all the elements of "all_energy_values" list
print(f'Total Energy: {Total_energy} kcal/mol')
```
